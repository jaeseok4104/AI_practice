```python
def f3(x,w):
    return (x-w)*x*(x+2)

plt.figure(figsize=(10,3))
plt.subplots_adjust(wspace=0.5, hspace=0.5)

x=np.linspace(-3,3,50)

for i in range(6):
    plt.subplot(2,3,i+1)
    plt.title(i+1)
    plt.plot(x, f3(x,i),'k')
    plt.grid(True)
plt.show()
```

```python
def f4(x0,x1):
    r = 2*x0**2+x1**2
    return r*np.exp(-r)

n_point = 30
x= np.linspace(-2,2,n_point)
y= np.linspace(-2,2,n_point)
z= np.zeros((n_point,n_point))

for m in range(len(x)):
    for n in range(len(y)):
        z[m,n]=f4(x[m],y[n])
        
plt.figure(figsize=(3.5,3))
plt.gray()
plt.pcolor(z)
plt.colorbar()
plt.show()
```
