```python
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline

def circle (x1,y1,r): 
    x=[] // list 선언
    y=[]
    
    for degree in range(0,360):
        x.append(x1+r*math.cos(math.radians(degree))) // x좌표 및 y좌표 구하기 
        y.append(y1+r*math.sin(math.radians(degree)))
    plt.figure(figsize=(10,10))
    plt.plot(x,y)
    plt.grid()
    plt.show()
    
circle(5,10,5)
```

```python
def rect(tau,x1):
    if (((x1%1)>(1-(tau/2)))or((x1%1)<(tau/2))):
        return 1.0
    else:
        return 0
    
x = np.linspace(0,5,500)
y=[]

for i in x:
    y.append(rect(0.5,i))
plt.plot(x,y)
plt.show()
```
